# Import PurpleAir and AirNow data archives

## Formatting and cleaning functions

```python
def lower_and_underscore(s):
    return str(s).strip().lower().replace(' ','_')

def import_archive_data(fname, season):
    df = pd.read_csv(fname)
    df['season'] = season
    df.columns = df.columns.to_series().apply(lower_and_underscore)
    return df
```

### Generate line plots for each sensor-season

```python
def plot_seasonal_trends(df_data, df_sensors):
    """Plot data for season, for each sensor separately"""
    
    nSensors = len(df_sensors.index)
    figure, axis = plt.subplots(nSensors)
    figure.set_figheight(10)

    dataMin = min(df_data['concentration'])
    dataMax = max(df_data['concentration'])
    buffer = 0.05*(dataMax - dataMin)
    
    colors = ['green', 'pink']
    for i, sensor in df_sensors.iterrows():
        # Get data corresponding to current sensor
        df_sensor = df_data.loc[(df_data['latitude'] == sensor['latitude']) &
                                (df_data['longitude'] == sensor['longitude'])]
        
        # Plot data for each season separately
        for j, season in enumerate(SEASONS):
            df_season = df_sensor.loc[df_sensor['season'] == season].copy()
            df_season.sort_values(by='datetime', inplace=True)
            
            axis[i].plot(df_season['datetime'], df_season['concentration'], color=colors[j])
        
        # Format plot
        axis[i].set_ylabel("Concentration\n(ug/m3)")
        axis[i].set_title(sensor['location'])
        axis[i].set_ylim([dataMin - buffer, dataMax + buffer])
        if i < nSensors-1:
            axis[i].set_xticklabels([])
        else:
            axis[i].xaxis.set_major_formatter(mdates.DateFormatter("%m/%d/%Y"))
            plt.xticks(rotation=60) 
```

## What is the correlation between sensors?

```python
def plot_heatmap(matrix, title='', vmin=0, vmax=1):
    """Plot correlation matrix or p-value matrix"""
    plt.figure()
    fig = sns.heatmap(matrix, cmap=sns.color_palette("Blues", as_cmap=True), vmin=0, vmax=1)
    plt.title(title)
    fig.xaxis.set_ticks_position("top")
```

```python
def sensor_correlation_matrix(df_data, df_sensors):
    """For each season, find the correlation between sensors.
    This requires interpolation: finding the overlapping time period for each pair of sensors
    and interpolating the concentration data with 2-hour frequency.
    This is done both for the raw data and for 24-hour rolling average data. 
    """
    
    nSensors = len(df_sensors.index)
    nSeasons = len(SEASONS)
    
    correlations_raw = np.zeros([nSeasons, nSensors, nSensors])
    correlations_rolling = np.zeros([nSeasons, nSensors, nSensors])
    
    pvalues_raw = np.zeros([nSeasons, nSensors, nSensors])
    pvalues_rolling = np.zeros([nSeasons, nSensors, nSensors])
    
    # For debugging/quality check: how many data points are included?
    nPointsArray = np.zeros([nSeasons, nSensors, nSensors])
    dateRangesArray = np.zeros([nSeasons, nSensors, nSensors])
    
    for i, season in enumerate(SEASONS):
        # Get data corresponding to current season
        df_season = df_data.loc[df_data['season'] == season].copy()
        df_season.sort_values(by='datetime', inplace=True)
        
        for j, sensor1 in df_sensors.iterrows():
            # Get data corresponding to current sensor
            df_sensor1 = df_season.loc[(df_season['latitude'] == sensor1['latitude']) &
                                       (df_season['longitude'] == sensor1['longitude'])]     
            
            for k, sensor2 in df_sensors.iterrows():
                # Correlate w/data for all sensors (including self)
                df_sensor2 = df_season.loc[(df_season['latitude'] == sensor2['latitude']) &
                                           (df_season['longitude'] == sensor2['longitude'])]
                
                if (len(df_sensor1.index) > 0) and (len(df_sensor2.index) > 0):
                    # Find overlapping time period and interpolate 
                    dtMin = max(min(df_sensor1['datetime']), min(df_sensor2['datetime']))
                    dtMax = min(max(df_sensor1['datetime']), max(df_sensor2['datetime']))
                    dtRange = pd.date_range(start=dtMin, end=dtMax, freq='2H')
                    
                    # For debugging/quality check: record date range/number of points used for correlation
                    nPointsArray[i, j, k] = len(dtRange)
                    dateRangesArray[i, j, k] = (dtMax-dtMin).days
                    
                    # Interpolate raw and rolling-average data
                    s1_interp_raw = np.interp(dtRange, df_sensor1['datetime'], df_sensor1['concentration'])
                    s2_interp_raw = np.interp(dtRange, df_sensor2['datetime'], df_sensor2['concentration'])
                    s1_interp_rolling = np.interp(dtRange, df_sensor1['datetime'], df_sensor1['concentration_rolling'])
                    s2_interp_rolling = np.interp(dtRange, df_sensor2['datetime'], df_sensor2['concentration_rolling'])

                    # Find Pearson correlation coefficients and add to correlation matrix
                    correlations_raw[i, j, k] = pd.Series(s1_interp_raw).corr(pd.Series(s2_interp_raw))
                    correlations_rolling[i, j, k] = pd.Series(s1_interp_rolling).corr(pd.Series(s2_interp_rolling))
                    
                    # Find p-value for relation between sensors
                    _, pvalues_raw[i, j, k] = stats.ttest_ind(s1_interp_raw, s2_interp_raw)
                    _, pvalues_rolling[i, j, k] = stats.ttest_ind(s1_interp_rolling, s2_interp_rolling)     

    if DEBUG:
        print(f"Date spans included in analysis ranged from {np.min(dateRangesArray)} days to {np.max(dateRangesArray)} days")
        print(f"Number of data points included for each sensor-pair correlation ranged from {np.min(nPointsArray)} points to {np.max(nPointsArray)} points")   
    
    correlations = {'raw': correlations_raw, 'rolling': correlations_rolling}
    pvalues = {'raw': pvalues_raw, 'rolling': pvalues_rolling}
    
    return correlations, pvalues
```

### Find correlation coefficients between AirNow sensors

```python
print(sensorCoords_an['location'])
correlations, pvalues = sensor_correlation_matrix(df_an_pm25, sensorCoords_an)
```

```python
# Plot correlation coefficient matrix
for i, season in enumerate(SEASONS):
    for key, data in correlations.items():
        plot_heatmap(data[i,:,:], title=f'Correlation coefficient, {key} data ({season})')     
```

```python
# Plot whether difference between sensors is statistically significant (p-value < 0.05)
for i, season in enumerate(SEASONS):
    for key, data in pvalues.items():
        isSignificant = np.where(data[i,:,:] < 0.05, 1, 0)
        plot_heatmap(isSignificant, title=f'p-value significance (is p < 0.5?), {key} data ({season})')  
```

## Generate time-series data comparing daily means between two sensors (same season)

```python
# sensor1 and sensor2 are specified by the pair (dataset, location) as in:
#   output_df = daily_comparison(('an', 'Cicero'), ('pa', '39th and Damen (NLEI)'), 'autumn')
def daily_comparison(sensor1, sensor2, season: str):
    # narrow scope to the given season
    s1 = sensor_data[sensor1[0]][sensor1[1]][lambda row: row['season'] == season].copy()
    s2 = sensor_data[sensor2[0]][sensor2[1]][lambda row: row['season'] == season].copy()
    # group by day, take the daily mean
    s1g = s1.groupby(pd.Grouper(key = 'datetime', freq = '1D')).mean()[['concentration']].rename(columns = {'concentration': sensor1[1]})
    s2g = s2.groupby(pd.Grouper(key = 'datetime', freq = '1D')).mean()[['concentration']].rename(columns = {'concentration': sensor2[1]})
    # strip down the datetime to just the date
    s1g.index = s1g.index.date
    s2g.index = s2g.index.date
    # merge the two sensors' data into a single dataframe, dropping any days on which either sensor has no data
    return s1g.join(s2g, how = 'inner')
#test_comp = daily_comparison(('an', 'Alsip'), ('an', 'Ashburn'), 'autumn')
#sns.scatterplot(data = test_comp, x = 'an_conc', y = 'pa_conc');
#sns.lineplot(data = test_comp);
```

### Display daily comparisons for each season

```python
for season in SEASONS:
    # get all sensors that have data for the given season
    locs_season = [loc for key in dfs.keys()
                    for loc in sensor_data[key].keys()
                    if not sensor_data[key][loc][lambda row: row['season'] == season].empty]
    nlocs = len(locs_season)
    fig, axs = plt.subplots(nlocs, nlocs)
    fig.set_figwidth(100)
    fig.set_figheight(100)
    fig.set_facecolor('white')
    for i, loc1 in enumerate(locs_season):
        for j, loc2 in enumerate(locs_season):
            if i == j:
                continue
            dataset1 = [key for key in sensor_data.keys() if loc1 in sensor_data[key].keys()][0]
            dataset2 = [key for key in sensor_data.keys() if loc2 in sensor_data[key].keys()][0]

            comp_df = daily_comparison((dataset1, loc1), (dataset2, loc2), season)

            corr_string = f'r = {comp_df.iloc[:, 0].corr(comp_df.iloc[:, 1]):.2f}'

            axs[i][j].plot(comp_df.index, comp_df.iloc[:, 0], label = comp_df.columns[0])
            axs[i][j].plot(comp_df.index, comp_df.iloc[:, 1], label = comp_df.columns[1])
            axs[i][j].plot([], [], ' ', label = corr_string)
            axs[i][j].tick_params('x', labelrotation = 60)
            axs[i][j].legend()
    fig.savefig('figures/compare_' + season + '.png')
```
